```
"""n_invigilator = 8
n_shift = 4
total_invigilator_demands = 20

# Demand Matrix
shift_demands = np.array([ 5, 4,  7, 4])
# Supply Matrix
invigilator_supply = np.array([ 2, 3, 3, 3, 3, 2, 2, 2])"""
n_invigilator = 115                                       # số lượng cán bộ coi thi
n_shift = 48                                               # số lượng ca thi cần coi
total_invigilator_demands = 1309
# Demand Matrix
# Mảng chứa tổng số cán bộ coi thi mà  mỗi ca thi cần 
shift_demands = np.array([ 29, 115,  14,   0,  15, 104,  15,   0,  14, 108,  23,   0,  14,
        92,  22,   0,  27,  52,  20,   0,  43,  39,   0,   0,  14,  53,
        32,   0,  12,  73,  30,   0,   7,  47,  32,   0,  16,  49,  38,
         0,  20,  34,  36,   0,  30,  38,   2,   0])

# Supply Matrix
# Mảng chứa tổng số ca thi mà cán bộ coi thi có thể coi
invigilator_supply = np.array([12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12,
       12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 12,
       12, 12, 12, 12, 12, 12, 12, 12, 12, 12, 11, 11, 11, 11, 11, 11, 11,
       11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11,
       11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11,
       11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11,
       11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11, 11])
```

```
model = LpProblem("Supply-Demand-Problem", LpMinimize) # Model optimize
```

```
# Tạo mảng một chiều chứa biến MAX 
maxxx=[i for i in range(1, n_invigilator+1)]
DV_variables = LpVariable.matrix("MAX", maxxx, cat = "Integer", lowBound= 0  )
maxx = np.array(DV_variables)
maxx
```

```
# Tạo mảng một chiều chứa biến MIN
minnn=[i for i in range(1, n_invigilator+1)]
DV_variables = LpVariable.matrix("MIN", minnn, cat = "Integer", lowBound= 1  )
minn = np.array(DV_variables)
minn
```

```
# Biến X_ có giá trị (0,1) biểu diễn cán bộ coi thi m có cần coi thi ca thi n hay không
# Tạo mảng hai chiều chứa biến trong 
variable_names = [str(i) for i in range(1, n_invigilator*n_shift+1)]
DV_variables = LpVariable.matrix("X", variable_names, cat = "Integer", lowBound= 0,upBound=1 )
allocation = np.array(DV_variables).reshape(115, 48)
print("Decision Variable/Allocation Matrix: ")
print(allocation)
```

```
# INPUT Hàm mục tiêu cần tối ưu
obj_func = lpSum(maxx[i]-minn[i] for i  in range(1, n_invigilator))
model +=  obj_func
print(model)
```

```
#Max Constraints
n_max=0
for i in range(n_invigilator):
    for j in range(n_shift):
        print(maxx[i]>=(j+1)*allocation[i][j]-1000*(1-allocation[i][j]))
        model += maxx[i]>=(j+1)*allocation[i][j]-1000*(1-allocation[i][j]), "Max Constraints " + str(n_max)
        n_max = n_max+1
```

```
#Min Constraints
n_min=0
for i in range(n_invigilator):
    for j in range(n_shift):
        print(minn[i]<=(j+1)*allocation[i][j]+1000*(1-allocation[i][j]))
        model +=minn[i]<=(j+1)*allocation[i][j]+1000*(1-allocation[i][j]), "Min Constraints " + str(n_min)
        n_min = n_min+1
```

```
#Supply Constraints
# Tổng hàng của mảng biến 
for i in range(n_invigilator):
    print(lpSum(allocation[i][j] for j in range(n_shift)) == invigilator_supply[i])
    model += lpSum(allocation[i][j] for j in range(n_shift)) == invigilator_supply[i] , "Supply Constraints " + str(i)
```

```
# Demand Constraints
# Tổng cột
for j in range(n_shift):
    print(lpSum(allocation[i][j] for i in range(n_invigilator)) == shift_demands[j])
    model += lpSum(allocation[i][j] for i in range(n_invigilator)) == shift_demands[j] , "Demand Constraints " + str(j)
```

```
model.writeLP("Supplydemandprob1.lp")
```

```
#model.solve()
model.solve(GLPK())

status =  LpStatus[model.status]

print(status)
```

```
print("Total Cost:", model.objective.value())

# Decision Variables tra trea
for v in model.variables():
    try:
        print(v.name,"=", v.value())
    except:
        print("error couldnt find value")
```
